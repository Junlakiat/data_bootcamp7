**HW IMDB movie top 50 & TV Show**

```r
library(rvest)
library(tidyverse)
```

IMDB top 50

```r
url <- "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"

movie_name <- url %>%
  read_html() %>%
  html_elements("h3.lister-item-header")  %>%
  html_text2()

rating <- url %>%
  read_html() %>%
  html_elements("div.ratings-imdb-rating")  %>%
  html_text2() %>%
  as.numeric()

imdb_top50 <- data.frame(movie_name, rating)

View(imdb_top50)

imdb_top50 %>%
  summarise(avg_rating = mean(rating),
                  min_rating = min(rating),
                  median_rating = median(rating),
                  max_rating = max(rating),
                  n= n())
```

IMDB top 10 rating TV Shows

```r
url <- "https://www.imdb.com/chart/toptv/?ref_=nv_td_tv250"

title <- url %>%
  read_html() %>%
  html_elements("td.titleColumn") %>%
  html_text2()

rating <- url %>%
  read_html() %>%
  html_elements("td.imdbRating") %>%
  html_text2() %>%
  as.numeric()

top250_tv_show <- data.frame(title, rating)

head(top250_tv_show, 10)

top250_tv_show %>%
  summarise(avg_rating = mean(rating),
            min_rating = min(rating),
            median_rating = median(rating),
            max_rating = max(rating),
            n= n())
```
